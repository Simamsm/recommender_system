```python
# create a function that gets a string as an input and returns the word embeddings of that string
```

```python
def get_embeddings(sentence):
  model = FlagModel('BAAI/bge-small-en-v1.5')
  embeddings = model.encode(sentence)astype(np.float32)
  return embeddings
```

```python
df['embed_description'] = df['description'].apply(get_embeddings)
```

```python
df
```

```python
embed_1 = get_embeddings('Hello!')
embed_2 = get_embeddings('Hello World!')
```

```python
# Calculate cosine similarity
similarity = cosine_similarity(embed_1.reshape(1, -1), embed_2.reshape(1, -1))
```

```python
similarity
```

```python
df.head()
```

```python
embed_query = get_embeddings('I am from Spain and I love soccer.')
```

```python
embed_query = get_embeddings('I love my family and I love romantic movies')
```

```python
similarities_list = []
for x in df['embed_description']:
  similarity = cosine_similarity(embed_query.reshape(1, -1), x.reshape(1, -1))
  similarity = similarity[0][0]
  similarities_list.append(similarity)

df['similarities'] = similarities_list
df.head()
```

```python
df = df.sort_values(by='similarities', ascending=False).reset_index(drop=True)
df
```
